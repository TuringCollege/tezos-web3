# Module 2 – Introduction to Tezos for developers
## Sprint 2 – Governance, use cases of Tezos and LIGO
## Part 2 – LIGO pt. 2 - CLI and transactions

In this part, we will start using the CLI to work with LIGO. Luckily, installing it has been made quite simple now if you’re using Docker. This will give you a good starting point for more complicated use cases with more complex smart contracts later on. In this part, you will also start writing contracts that can make transactions or even interact with other smart contracts!


## Part Tasks
- Prepare to run LIGO via CLI. We recommend the Dockerised installation. Note that if you get “missing subcommand for command ligo”, everything is working and you are ready to use the compiler. Alias the docker command as recommended in the link however, as otherwise it will be very bothersome to use.
  - https://ligolang.org/docs/intro/installation 
- You may also use the following page for reference when running various commands or deploying your smart contracts:
  - https://opentezos.com/ligo/deploy-a-contract 
- Go to chapter 8 of [Ligo Academy](https://academy.ligolang.org/js/chapter-tuples), save the solution code there and run the suggested command to interpret the code. Make sure the result of the function is returned.
- Go through Chapters 14-21 of [Tezos Academy](https://academy.ligolang.org/js/chapter-built-ins)
  - In chapter 16 (Built-ins), try to run the given code as suggested using installed LIGO. It should fail – find out why and make the code run without any warnings (https://ligolang.org/docs/reference/current-reference/ might be useful). Next, change the address and see if the expected error is returned. If not, make the required changes needed.

- Koans 6-7 of https://github.com/ligolang/ligo-koans
