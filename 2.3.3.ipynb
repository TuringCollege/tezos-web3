## Sprint 3 – Tezos in-depth
## Part 3 – More Michelson


Let’s continue learning about how Michelson works. By the end of this part, you will have a good basic understanding of this language. In case you need to use it more extensively, you should be able to use this knowledge as a strong  basis for further, deeper learning. If not, it is still going to be interesting to see how the high-level languages actually follow the ideas and structure of Michelson quite  strongly.

## Direction for further research
- Try to find an actual deployed smart contract and understand how at least a part of it (maybe one entry point) works. You can search for popular dapps (which have underlying smart contracts) here: https://tzkt.io/dapps 
- Write a smart contract in any of your preferred high-level language (LIGO/Archetype/Smartpy), transpile it to Michelson. Check whether you fully understand the generated code. Do you think it is possible to write the program in a more optimized Michelson code than what was generated?
