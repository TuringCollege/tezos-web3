## Part 1 – Liquid Proof of Stake in Tezos
A unique and interesting part of Tezos is the Liquid Proof of Stake consensus mechanism. Interestingly, the on-chain governance allowed it to be changed significantly since its original version. Many core constants have been changed, such as the time to mine the blocks and even the whole algorithm was swapped out from Emmy to TenderBake. In this part, you will look deeply into how the current consensus algorithm works. Understanding it will not only mean that you are knowledgeable of one of the core aspects of Tezos but also allow you to (if you desire) get into baking or delegating fully understanding the system.
