## Sprint 2 – Governance, use cases of Tezos and LIGO
## Part 4 – LIGO pt.3 - advanced concepts

You have now had quite some practice using LIGO. In this part, we will go into some final, more advanced concepts. Don’t be surprised if this part takes you more time to complete. If you see that it’s especially time consuming, feel free to move on to the practical project (the next part) after some time even though there are still some Koans or Academy chapters to fully understand. Go back to them later on – either when you see that they are needed for the project or if you simply have time. It’s quite common to read something that seems difficult and not understand it, leave it, then come back to it a day later and suddenly realize that everything makes sense now!
