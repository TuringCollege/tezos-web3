## Sprint 2 – Governance, use cases of Tezos and LIGO
## Part 3 – Use cases and notable projects

While developing smart contracts and working with the blockchain can be an interesting activity in itself, it’s important to start thinking about how to make all of it into something useful that people will want to use. In this part, we will look into what kind of use cases are being explored and implemented already. Most of it will focus on the Tezos blockchain, but we will end with a resource which is not specific to a particular blockchain, instead simply listing many possible use cases that you can explore. Many of the resources in this part will be links to specific projects. In module 1 of this course, you have learned how to critically and deeply evaluate various projects – it’s time to practice these skills once again!

  <BR> It is one the successors to https://www.hicetnunc.xyz/, the first NFT marketplace on Tezos where artists could mint and sell NFTs. Hic Et Nunc (or short Hen) was started by one person and was a massive success, but was eventually discontinued. Most of the original hen community has since organized themselves under the name Teia. Has a very active Discord community.
