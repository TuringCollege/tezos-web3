# Module 2 – Introduction to Tezos for developers
## Sprint 3 – Tezos in-depth
## Part 4 – Advanced Tezos Features - Tokens, Tickets, Oracles


The Tezos ecosystem provides many different functionalities and features that you can use to develop advanced smart contracts for various different use cases. While it’s not worth going extremely in-depth into each one of them right now, when you’re just beginning to get into working with Tezos, it’s still useful to get an idea of these features. This will allow you to better understand what potential things you can create with Tezos and what topics to go much deeper into for specific implementations. 
  The part will also end with a suggestion for further research that we strongly recommend trying out – finding documentation projects that you can make suggestions to. This can of course be a great way to start becoming more well-known in the community and at the same time a way to make yourself pay closer attention to the resources that you read.

## Direction for further research
- Would you be able to make any kind of a GET/POST request with your smart contract with the help of oracles? The following resource can be a good starting point to look through 
- Transaction rollups have an extremely long finality time. What happens if a rollup is deemed invalid after many blocks and many subsequent transactions that depend on the transactions inside that rollup?
Direction for further research: besides bug bounties mentioned in this sprint, an even easier way to contribute to the Tezos ecosystem is to help with documentation projects. A perfect example is a website you are familiar with - OpenTezos. It is still being developed, but already has some outdated information that can be improved. They also have instructions for contributing: https://opentezos.com/contribute/opentezos. Proposing updates here can be a great way to start making a reputation for yourself in the Tezos community, which is crucial for getting better job opportunities. You can also look for other similar projects and ask if they would like similar help (and in what format).
