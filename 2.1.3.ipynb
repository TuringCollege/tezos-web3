# Module 2 – Introduction to Tezos for developers
## Sprint 1 – Getting started with Tezos
## Part 3 – Governance in Tezos

## About this Part
You have now heard quite a bit about the self-governance in Tezos and experienced in practice how there are different amendments (similar to versions of a program) that you can interact with. In this part we will look in depth into how the governance process works and how the new amendments get implemented. This will allow you to understand one of the most unique aspects of the Tezos blockchain. After understanding how it works, to get a more nuanced view, you will also read some opinion articles about cases where such a governance system might not work and what the dangers of it are.
